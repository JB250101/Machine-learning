**1. The relationship between the penguin's bill length and bill depth for each species.** (5 marks)

```python
## TODO: Check your attributes, make a graph using Matplotlib or Seaborn to show the relationship needed
sns.set(style='darkgrid')

plt.figure(figsize=(11,7))
scatter = sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', hue = 'species', data= penguins)

plt.title("Relationship between Penguin Bill Length and Bill Depth by Species")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")

plt.legend(title='Species')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Create a box plot for bill length by species
plt.subplot(1, 2, 1)  # Create a 1x2 subplot, first plot
sns.boxplot(x='species', y='bill_length_mm', data=penguins,hue= 'species', palette='Set2')
plt.title('Box Plot of Bill Length by Species')
plt.xlabel('Species')
plt.ylabel('Bill Length (mm)')

# Box Plot for Bill Depth
plt.subplot(1, 2, 2)  # Second plot
sns.boxplot(x='species', y='bill_depth_mm', data=penguins,hue= 'species', palette='Set2')
plt.title('Box Plot of Bill Depth by Species')
plt.xlabel('Species')
plt.ylabel('Bill Depth (mm)')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(12,6))

#plot of bill_length_mm
plt.subplot(1,2,1)
sns.kdeplot(data=penguins, x='bill_length_mm', hue='species', fill= None, palette='Set2')

plt.title('Density Plot of Bill Length by Species')

#plot of bill_depth_mm
plt.subplot(1,2,2)
sns.kdeplot(data=penguins, x='bill_depth_mm', hue='species', fill= None, palette='Set2')

plt.title('Density Plot of Bill depth by Species')

# Show the plot
plt.tight_layout()
plt.show()

```

**2. Suitable graph of the mean BMI (BMI not body_mass; BMI Equation is: weight_in_kg/height^2) of female and male penguins. (Assume the height is the flipper length)**  (5 marks)


```python
# BMI [(body_mass_g) /1000) / (flipper_length_mm / 1000)^2)]
penguins['BMI']= penguins['body_mass_g'] / 1000 / ((penguins['flipper_length_mm'] / 1000)**2)

# Since we need mean BMI of female and male penguins we will group them by 'sex'
mean_BMI = penguins.groupby('sex')['BMI'].mean().reset_index()

# Bar plot of mean  BMI, since its categorical
plt.figure(figsize=(8, 5))
sns.barplot(x='sex', y='BMI', data= mean_BMI, palette='Set2')
plt.title('Mean BMI of Male and Female Penguins')
plt.xlabel('Sex')
plt.ylabel('Mean BMI')
plt.ylim(0, mean_BMI['BMI'].max() + 1)  # Add some space above the tallest bar

# Show the plot
plt.show()
```

 **What do you observe from the graph?** (2 marks)

Observations from the Graph
1. Comparison of Mean BMI:

    * The bar plot clearly shows the mean BMI for both male and female penguins. From visual inspection, it appears that the mean BMI for male penguins is higher than that of female penguins.
    * This difference may suggest that male penguins, on average, possess greater body mass relative to their height (flipper length) compared to females.
2. Magnitude:

    * Both values are significant, but with male penguins generally having a noticeably higher mean BMI. This may indicate different growth patterns, nutrition, or reproductive strategies between the sexes.
   

**Why did you choose this type of graph?** (2 marks)

1. Clarity in Comparison:

    * A bar graph is particularly effective for comparing categorical data, in this case, the mean BMI of two distinct categories (male and female). It allows for easy visualization of differences in means between the groups at a glance.
2. Simplicity and Directness:

    * The bar plot provides a straightforward representation of the data, making it easy for observers to understand differences in mean BMI without having to interpret complex data points. Each bar serves as a visual summary of the mean for each sex, enhancing the communication of results.
