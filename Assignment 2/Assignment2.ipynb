```python
import requests
import pandas as pd
from itertools import cycle
from collections import defaultdict

with requests.get("http://lib.stat.cmu.edu/datasets/boston") as response:
    response.raise_for_status()
    lines = response.text.splitlines()
    data = defaultdict(list)
    columns = [line.split()[0] for line in lines[7:21]]
    keys = cycle(columns)
    for line in lines[22:]:
        for token in line.split():
            data[next(keys)].append(token)
    boston = pd.DataFrame(data)
 
boston.head(-1)
```

# Z-score statiscal analysis

```python
import numpy as np
from scipy import stats

bostonf = boston.astype(float)

# Z-Score Method
z_scores = np.abs(stats.zscore(bostonf))
z_outliers = (z_scores > 3).any(axis=1)

# IQR Method
Q1 = bostonf.quantile(0.25)
Q3 = bostonf.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
iqr_outliers = (bostonf < lower_bound) | (bostonf > upper_bound)

# Identify and print outliers
print("Z-score Outliers:")
print(bostonf[z_outliers])

print("IQR Outliers:")
print(bostonf[iqr_outliers.any(axis=1)])
```

```python
# Clean the dataset: combine Z-score and IQR outlier detections
combined_outliers = z_outliers | iqr_outliers.any(axis=1)

# Create cleaned dataset
boston_cleaned = bostonf[~combined_outliers]

# Display cleaned data
print("Cleaned Data:")
boston_cleaned.head(-1)
```

#### Question 7: Implementing the Local Outlier Factor (LOF) Function (25)


- Local outlier factor is a function that labels outlier based on 2 predefined parameters, the first one is t **k** for k-Nearest-Neighbours. The second one is the threshold of outliers which could be set arbitrary. However, in this coding task you are required to BOX-PLOTS to determine the threshold for the local outlier factor and label your outliers.
- You are expected to implement different functions to reach the target which will be defined below.


```python
def calculate_point_to_point_distance(point_a, point_b, distance_approach="Euclidean"):
    ## TODO: This function takes as an input two points (where each point is a pair of (x,y)) and returns the distance between the points.
    ## The distance is calculate based on given distance_approach
    ## You need to handle the Euclidean and Manhattan Distance Approaches in your code
    if distance_approach == "Euclidean":
        return np.sqrt(np.sum((np.array(point_a) - np.array(point_b))**2))
    elif distance_approach == "Manhattan":
        return np.sum(np.abs(np.array(point_a) - np.array(point_b)))
    else:
        raise ValueError("Invalid distance approach. Use 'Euclidean' or 'Manhattan'.")

# Function to get the k-th nearest neighbor and its reachability
def get_k_nearest_neighbors(point, list_of_points, K, distance_approach):
    distances = [calculate_point_to_point_distance(point, other_point, distance_approach) for other_point in list_of_points]
    sorted_indices = np.argsort(distances)
    
    neighbors = sorted_indices[1:K + 1]  # Exclude the point itself
    Nd = distances[neighbors[-1]]  # Distance to the K-th nearest neighbor
    
    # Points within the neighborhood of distance Nd
    neighborhood = [i for i in range(len(list_of_points)) if distances[i] <= Nd]  
    return neighbors, Nd, neighborhood

def get_points_lof(list_of_points, K=2, distance_approach = "Euclidean"):
    ## TODO: This function takes a list of points and returns a dictionary
    ## The dictionary is in format {point: LOF, ...}
    ## You can use as much helper functions as you'd like
    LOF_scores = {}

    for i, point in enumerate(list_of_points):
        neighbors, Nd, neighborhood = get_k_nearest_neighbors(point, list_of_points, K, distance_approach)

        # Initialize reach distances
        reach_distances = []
        
        # Calculate reach distance for each neighbor in the neighborhood
        for idx in neighborhood:
            distance = calculate_point_to_point_distance(point, list_of_points[idx], distance_approach)
            reach_distance = max(distance, Nd)
            reach_distances.append(reach_distance)

        # Calculate Local Reachability Density (LRD)
        LRD = len(neighborhood) / np.sum(reach_distances) if np.sum(reach_distances) > 0 else 0
        
        # Calculate LOF
        LOF_sum = 0
        for idx in neighborhood:
            neighbor_Nd = get_k_nearest_neighbors(list_of_points[idx], list_of_points, K, distance_approach)[1]
            neighbor_LOF_score = len(get_k_nearest_neighbors(list_of_points[idx], list_of_points, K, distance_approach)[2]) / neighbor_Nd if neighbor_Nd > 0 else 0
            LOF_sum += neighbor_LOF_score

        LOF = LOF_sum / (len(neighborhood) ** 2) if len(neighborhood) > 0 else 0
        
        LOF_scores[tuple(point)] = LOF  # Store the LOF score using a tuple as the key       

    return LOF_scores

def get_outliers(list_of_points, K=2, distance_approach="Euclidean"):
    ## TODO: This function uses the get_points_lof to get the LOF scores, and uses is to build a box-plot
    ## Using the box-plot points are labelled as outliers are returned to the user
    ## Example: if point (2,1) and (3,2) are labelled as outliers using box-plot
    ## Then you will return the example given in the place-holder below
    # Get LOF scores
    LOF_scores = get_points_lof(list_of_points, K, distance_approach)
    
    # Extract LOF values
    lof_values = np.array(list(LOF_scores.values()))
    
    # Create a box plot to identify outliers
    plt.figure(figsize=(10, 5))
    plt.boxplot(lof_values)
    plt.title('Box Plot of LOF Values')
    plt.ylabel('LOF Score')
    plt.show()
    
    # Determine outlier cutoff (e.g., points above 1.5 * IQR)
    Q1 = np.percentile(lof_values, 25)
    Q3 = np.percentile(lof_values, 75)
    IQR = Q3 - Q1
    outlier_cutoff = Q3 + 1.5 * IQR
    
    
    # Identify outliers based on LOF scores
    outliers = [point for point, score in LOF_scores.items() if score > outlier_cutoff]
    
    return outliers
```

##### To calculate the LOF you need to apply the following steps:
 1. Calculate the distance between each pair of points p1,p2
 2. Find your Kth nearest neighbor point (x) from current point (p):
    - After finding it, remember the distance *Nd* which is the distance from p to x, and save a set of elements which are in a neighborhood of *Nd* around point p.
    - Save the length of this set of elements.
 3. Calculate ReachDistance(x,p)=max(Nd_x, Distance(x,p))
 3. Calculate LRD (Local Reachability Distance) where LRD_p = length_of_set_p/Sum[ReachDistance(x,p)] (for each x in set of elements of point p)
 4. Calculate LOF (Local Outlier Factor) where LOF_p = {Sum[LRD_x] (for each x in set of elements of point p) * Sum[RD(x,p)] (for each x in set of elements of point p)} / length_of_set_p^2
